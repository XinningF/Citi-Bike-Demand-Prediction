```python
import pandas as pd
%matplotlib inline
```

```python
# Load weather data
weather1 = pd.read_csv('./Data/Weather/3326992.csv')
weather2 = pd.read_csv('./Data/Weather/3326994.csv')
weather_data = pd.concat([weather1, weather2], axis = 0)

# weather_data.info()
```

AWND - Average daily wind speed (meters per second or miles per hour as per user preference)

DAPR - Number of days included in the multiday precipitation total (MDPR)

DASF - Number of days included in the multiday snowfall total (MDSF)

MDPR - Multiday precipitation total (mm or inches as per user preference; use with DAPR and DWPR, if available)

MDSF - Multiday snowfall total (mm or inches as per user preference)

SNWD -  Snow depth (mm or inches as per user preference, inches on Daily Form pdf file)

```python
weather_data = weather_data.drop(["WT01", "WT02", "WT03", "WT04", "WT05", "WT06", "WT08", "WT09", "WT11"], axis = 1)
weather_data.info()
weather_data.head()
```

```python
# weather_data["DATE"][0] == "2022-01-01"
weather_data["DATE"] = pd.to_datetime(weather_data["DATE"])
weather_data = weather_data.drop(["STATION", "NAME", "DAPR", "DASF", "MDSF", "MDPR"], axis = 1)
```

```python
weather_data_cleaned = pd.DataFrame({"Date": [], "AWND": [], "PRCP": [], "SNOW": [], "SNWD": [], "TAVG": [], "TMAX": [], "TMIN": []})
for d in pd.date_range(start='4/1/2021', end='4/30/2023'):
    date_data = weather_data.loc[weather_data["DATE"] == pd.to_datetime(d)]
    data_to_insert = date_data.mean(skipna = True)
    data_to_insert["Date"] = pd.to_datetime(d)
    weather_data_cleaned = weather_data_cleaned.append(data_to_insert, ignore_index = True)
```

```python
weather_data_cleaned.to_csv("./Data/Weather/cleaned_weather.csv", index = False)
```
